```python
# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_9949.csv'
df = pd.read_csv(file_path, skiprows=4)

# Filter the data for China
df_china = df[df['Country Name'] == 'China']

# Extract the years and corresponding population values for China
years = df_china.columns[4:-1]  # Select columns representing years (skip non-year columns)
population = df_china.iloc[0, 4:-1].values  # Get the population values for China

# Convert the year columns to numeric for proper plotting
years = pd.to_numeric(years)

# Create a bar chart for China's population over time
plt.figure(figsize=(10, 6))
plt.bar(years, population, color='olive')
plt.title('Population in China (1960-2023)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_9949.csv'
df = pd.read_csv(file_path, skiprows=4)

# Filter the data for China
df_china = df[df['Country Name'] == 'China']

# Extract the population values for China
population = df_china.iloc[0, 4:-1].values

# Create a histogram with a density curve for China's population distribution
plt.figure(figsize=(10, 6))
sns.histplot(population, bins=10, kde=True, color='b', edgecolor='black', alpha=0.7)

# Adding title and labels
plt.title('Population Distribution in China (1960-2023)', fontsize=16)
plt.xlabel('Population', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Improved grid style
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
